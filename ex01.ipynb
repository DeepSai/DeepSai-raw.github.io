1. 读取一张jpg图片
2. 将每个pixel的rgb值减半（eg:(122,224,225)=>(61,112,112)）,再将图片输出
3. rgb数值无条件舍去小数点

```python
# 导入用到的包
import math
from PIL import Image
```

```python
# 读取一张jpg图片
im = Image.open("pic.jpg")
im
```

```python
# 查看相关属性
im.format, im.size, im.mode
```

### 方法一
The Python Imaging Library also allows you to work with the individual bands of an multi-band image, such as an RGB image. The split method creates a set of new images, each containing one band from the original multi-band image. The merge function takes a mode and a tuple of images, and combines them into a new image.

>tip: Note that for a single-band image, split returns the image itself. To work with individual colour bands, you may want to convert the image to “RGB” first.

```python
r, g, b = im.split()

rr = r.point(lambda i: math.floor(i*0.5))
gg = g.point(lambda i: math.floor(i*0.5))
bb = b.point(lambda i: math.floor(i*0.5))

out = Image.merge("RGB", (rr, gg, bb))
out
```

### 方法二
The point method can be used to translate the pixel values of an image (e.g. image contrast manipulation). In most cases, a function object expecting one argument can be passed to the this method. Each pixel is processed according to that function:

`out = im.point(lambda i: i * 1.2)`  \# multiply each pixel by 1.2

```python
out = im.point(lambda i: math.floor(i*0.5))
out
```
